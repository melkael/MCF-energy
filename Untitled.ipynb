# Results with first instances generated

```python
import matplotlib.pyplot as plt
import pandas as pd

df_res = pd.read_csv('./MCF_min_resource/results/agis.csv')
energy_res = list(df_res['energy'])

df_cong = pd.read_csv('./MCF_min_congestion/results/agis.csv')
energy_cong = list(df_cong['energy'])

df_energy = pd.read_csv('./MCF_energy/results/agis.csv')
energy_ener = list(df_energy["energy"])
```

```python
plt.plot(energy_res, label="Resource Minimization")
plt.plot(energy_cong, label="Congestion Minimization", linestyle="dotted")
plt.plot(energy_ener, label="Energy Minimization", linestyle="--")

plt.xlabel("Number of commodities")
plt.ylabel("Energy consumption (watts)")
plt.title("Energy consumption with respect to the number of commodities")
plt.legend()
```

```python
cong_res = list(df_res['congestion'])
cong_cong = list(df_cong['congestion'])
cong_ener = list(df_energy["congestion"])

plt.clf()

plt.plot(cong_res, label="Resource Minimization")
plt.plot(cong_cong, label="Congestion Minimization", linestyle="dotted")
plt.plot(cong_ener, label="Energy Minimization", linestyle="--")

plt.xlabel("Number of commodities")
plt.ylabel("Congestion")
plt.title("Congestion with respect to the number of commodities")
plt.legend()
```

```python
res_res = list(df_res['resrouces'])
res_cong = list(df_cong['resouces'])
res_ener = list(df_energy["resouces"])

plt.clf()

plt.plot(res_res, label="Resource Minimization")
plt.plot(res_cong, label="Congestion Minimization", linestyle="dotted")
plt.plot(res_ener, label="Energy Minimization", linestyle="--")

plt.xlabel("Number of commodities")
plt.ylabel("Resource consumption")
plt.title("Resource consumption with respect to the number of commodities")
plt.legend()
```

```python
res_res = list(df_res['num_nodes'])
res_cong = list(df_cong['num_nodes'])
res_ener = list(df_energy["num_nodes"])

plt.clf()

plt.plot(res_res, label="Resource Minimization")
plt.plot(res_cong, label="Congestion Minimization", linestyle="dotted")
plt.plot(res_ener, label="Energy Minimization", linestyle="--")

plt.xlabel("Number of commodities")
plt.ylabel("Number of branch-and-bound nodes")
plt.yscale("log")
plt.title("Number of B-n-B nodes explored with respect to the number of commodities")
plt.legend()
```
